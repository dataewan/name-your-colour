```python
from IPython.display import display_svg


def display_colour(r, g, b):
    colour = "#%02x%02x%02x" % (r, g, b)
    display_svg("""
    <svg x="100" style="width: 100%; height: 100%;">
        <rect height="80" style="fill: {colour}" width="80" x="10" y="10"></rect>
    </svg>
    """.format(colour=colour), raw=True)
    
display_colour(22, 241, 103)
```

```python
answers = pd.read_sql("SELECT * FROM answers", con)
```

```python
for i, row in answers.sample(5).iterrows():
    print(row.colorname)
    display_colour(row.r, row.g, row.b)
```

```python
h = "#f4555d"


def hex2triple(hexval):
    __ = hexval.lstrip('#')
    return tuple(int(__[i:i+2], 16) for i in (0, 2 ,4))

display_colour(*hex2triple(h))
```

```python
answers.head()
```

```python
import math
from collections import namedtuple

point = namedtuple('point', 'r g b')
```

```python
p = point(*hex2triple(h))

def distanceto(row):
    return math.sqrt((row.r - p.r)**2 +
                     (row.g - p.g)**2 +
                     (row.b - p.b)**2)
```

```python
%time
answers_w_distances = (
    answers
    .sample(100000)
    .pipe(lambda x: x.assign(dist=x.apply(distanceto, axis=1)))
)
```

Do some nearest neighbours checking to see what that looks like.

```python
answers_w_distances.sort_values('dist')
```
